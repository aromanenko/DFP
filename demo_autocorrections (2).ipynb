# Generate Data
## DISACC_DISAGG_HYBRID_FORECAST, FORECAST_FLAG

```python
def generate_data(config_file : dict, 
                  IB_HIST_START_DT : datetime.datetime, 
                  IB_HIST_END_DT : datetime.datetime, 
                  hierarchies : dict,
                  PROMO : pd.DataFrame) -> (pd.DataFrame, pd.DataFrame, pd.DataFrame, pd.DataFrame):
    """
    Function generating input data
    
    Parameters
    ----------
    config_file : dict
        Configuration parameters used within the step
    IB_HIST_END_DT
        Last known date (i.e. sales and stock information is known)
    hierarchies
        Dictionary containg matches of key names with the relevant hierarchical tables
       
    Returns
    -------
    pd.DataFrame
        ACC_AGG_HYBRID_FORECAST_ used as input data of the algorithm
    pd.DataFrame
        DEMAND_RESTORED_ Demand information regarding 
        the past till last known day of the history.
    pd.DataFrame
        FORECAST_FLAG used as input data of the algorithm
    pd.DataFrame    
        PRE_ABT General table for ML and VF ABTs preparation
        used as input data of the algorithm
    """
    
    freq = config_file['out_time_lvl'][0]
    timerange = pd.date_range(IB_HIST_END_DT, IB_HIST_END_DT + datetime.timedelta(weeks=52), freq=freq)
    timerange += datetime.timedelta(1)
    
    ACC_AGG_HYBRID_FORECAST_ = pd.DataFrame(timerange, columns=['period_dt'])
    
    for key in ['product', 'location', 'customer', 'distr_channel']:
        column_name = f"{key}_lvl_id{config_file[f'vf_{key}_lvl']}"
        keys_df = pd.DataFrame(hierarchies[key][column_name]).drop_duplicates()
        ACC_AGG_HYBRID_FORECAST_ = pd.merge(ACC_AGG_HYBRID_FORECAST_, keys_df, 'cross')

    ACC_AGG_HYBRID_FORECAST_['segment_name'] = "Name of segment " + ACC_AGG_HYBRID_FORECAST_.index.astype(str)
    ACC_AGG_HYBRID_FORECAST_['vf_forecast_value'] = np.abs(np.random.normal(500, 300, 
                                                                            ACC_AGG_HYBRID_FORECAST_.shape[0]))
    ACC_AGG_HYBRID_FORECAST_['demand_type'] = np.random.choice(['promo', 'regular'], 
                                                               ACC_AGG_HYBRID_FORECAST_.shape[0])
    ACC_AGG_HYBRID_FORECAST_['assortment_type'] = np.random.choice(['new', 'old'], 
                                                               ACC_AGG_HYBRID_FORECAST_.shape[0])
    ACC_AGG_HYBRID_FORECAST_['ml_forecast_value'] = np.abs(np.random.normal(500, 300, 
                                                                            ACC_AGG_HYBRID_FORECAST_.shape[0]))
    ACC_AGG_HYBRID_FORECAST_['hybrid_forecast_value'] = np.abs(np.random.normal(500, 300, 
                                                                            ACC_AGG_HYBRID_FORECAST_.shape[0]))
    ACC_AGG_HYBRID_FORECAST_['ensemble_forecast_value'] = (ACC_AGG_HYBRID_FORECAST_['ml_forecast_value'] + 
                                                           ACC_AGG_HYBRID_FORECAST_['vf_forecast_value']) / 2
    ACC_AGG_HYBRID_FORECAST_['forecast_source'] = np.random.choice(['vf', 'ml'], 
                                                               ACC_AGG_HYBRID_FORECAST_.shape[0])

    
    DEMAND_RESTORED = pd.DataFrame()
    DEMAND_RESTORED['period_dt'] = pd.date_range(IB_HIST_START_DT, IB_HIST_END_DT)
    DEMAND_RESTORED = pd.merge(DEMAND_RESTORED, PRODUCT['product_id'], 'cross')
    DEMAND_RESTORED = pd.merge(DEMAND_RESTORED, LOCATION['location_id'], 'cross')
    DEMAND_RESTORED = pd.merge(DEMAND_RESTORED, CUSTOMER['customer_id'], 'cross')
    DEMAND_RESTORED = pd.merge(DEMAND_RESTORED, DISTR_CHANNEL['distr_channel_id'], 'cross')
    
    DEMAND_RESTORED['tgt_qty'] = abs(np.random.normal(500, 300, DEMAND_RESTORED.shape[0]))
    DEMAND_RESTORED['tgt_qty_r'] = DEMAND_RESTORED['tgt_qty'] + np.random.normal(50, 30, DEMAND_RESTORED.shape[0])
    DEMAND_RESTORED['stock_qty'] = abs(np.random.normal(500, 300, DEMAND_RESTORED.shape[0]))
    DEMAND_RESTORED['promo_flg'] = np.random.choice([0, 1], p = [0.8, 0.2], size = DEMAND_RESTORED.shape[0])
    DEMAND_RESTORED['promo_type'] = np.nan
    DEMAND_RESTORED.loc[DEMAND_RESTORED['promo_flg'] == 1, 'promo_type'] = np.random.choice(
        PROMO['promo_type'].tolist(), size=(DEMAND_RESTORED['promo_flg'] == 1).sum()
    )
    DEMAND_RESTORED = pd.merge(DEMAND_RESTORED, PROMO[['promo_type']], on='promo_type', how='left')
    DEMAND_RESTORED['deficit_flg1'] = np.random.choice([0, 1], p = [0.8, 0.2], size = DEMAND_RESTORED.shape[0])
    DEMAND_RESTORED['deficit_flg2'] = np.random.choice([0, 1], p = [0.8, 0.2], size = DEMAND_RESTORED.shape[0])
    
    FORECAST_FLAG = pd.DataFrame()
    FORECAST_FLAG['period_start_dt'] = [IB_HIST_START_DT]
    FORECAST_FLAG['period_end_dt'] = [IB_HIST_END_DT]

    for key in hierarchies:
        key_df = hierarchies[key][f"{key}_id"]
        FORECAST_FLAG = pd.merge(FORECAST_FLAG, key_df, 'cross')
    FORECAST_FLAG['status'] = np.random.choice(['active', 'blocked', 'out-of-sale'], size = FORECAST_FLAG.shape[0])

    PRE_ABT = pd.DataFrame()
    PRE_ABT['period_dt'] = pd.date_range(IB_HIST_START_DT, IB_HIST_END_DT)
    for key in hierarchies:
        key_df = hierarchies[key][f"{key}_id"]
        PRE_ABT = pd.merge(PRE_ABT, key_df, 'cross')
    PRE_ABT['tgt_qty'] = abs(np.random.normal(500, 300, PRE_ABT.shape[0]))
    PRE_ABT['tgt_qty_r'] = PRE_ABT['tgt_qty'] + np.random.normal(50, 30, PRE_ABT.shape[0])
    PRE_ABT['stock_qty'] = abs(np.random.normal(500, 300, PRE_ABT.shape[0]))
    PRE_ABT['deficit_flg1'] = np.random.choice([0, 1], p = [0.8, 0.2], size = PRE_ABT.shape[0])
    PRE_ABT['deficit_flg2'] = np.random.choice([0, 1], p = [0.8, 0.2], size = PRE_ABT.shape[0])
    PRE_ABT['promo_flg'] = np.random.choice([0, 1], p = [0.8, 0.2], size = PRE_ABT.shape[0])
    keys = ['product_id', 'location_id', 'customer_id', 'distr_channel_id']
    PRE_ABT = pd.merge(PRE_ABT, PROMO[keys + ['period_start_dt', 'period_end_dt', 'promo_price']], 
                  on=keys, how='left', suffixes=['', '_promo'])
    PRE_ABT['num_autorization'] = np.random.choice([0, 1], p = [0.8, 0.2], size = PRE_ABT.shape[0])
    PRE_ABT['promo_type'] = np.nan
    PRE_ABT.loc[(DEMAND_RESTORED['product_id'] == PROMO['product_id'][0]) &
                     (DEMAND_RESTORED['location_id'] == PROMO['location_id'][0]) &
                     (DEMAND_RESTORED['period_dt'] == PROMO['period_start_dt'][0]) &
                     (DEMAND_RESTORED['distr_channel_id'] == PROMO['distr_channel_id'][0]), 'promo_flg'] = 1  
    PRE_ABT.loc[PRE_ABT['promo_flg'] == 1, 'promo_type'] = np.random.choice(
        PROMO['promo_type'].tolist(), size=(PRE_ABT['promo_flg'] == 1).sum()
    )
    PRE_ABT = pd.merge(PRE_ABT, PROMO[['promo_type']], on='promo_type', how='left')
#     PRE_ABT['price_act']
#     PRE_ABT['price_reg']
    
    return ACC_AGG_HYBRID_FORECAST_, DEMAND_RESTORED, FORECAST_FLAG, PRE_ABT
    
```

```python
ACC_AGG_HYBRID_FORECAST_, DEMAND_RESTORED, FORECAST_FLAG, PRE_ABT = generate_data(config_file, 
                                                                                  IB_HIST_START_DT, 
                                                                                  IB_HIST_END_DT, 
                                                                                  hierarchies, 
                                                                                  PROMO)
```

```python
def delete_forecast_inactive_period(DISACC_DISAGG_HYBRID_FORECAST : pd.DataFrame, 
                                    FORECAST_FLAG : pd.DataFrame,
                                    hierarchies : dict,
                                    PRODUCT : pd.DataFrame,
                                    LOCATION : pd.DataFrame,
                                    CUSTOMER : pd.DataFrame,
                                    DISTR_CHANNEL : pd.DataFrame) -> (pd.DataFrame):
    """
    A case when the time series that should not be used in forecasting process, 
    has non-zero forecast for the forecast period. 
    The forecast should be fully cleared from these values.a
    
    Inputs
    ----------
    config_file : dict
        Configuration parameters used within the step
    out_time_lvl : string
            Required time granularity level
            Possible values:
            D - days
            W - weeks (default starting from Sunday)
            W-MON/W-TUE/.../W-SUN - weeks, starting from specified day of week
            M - months
    DISACC_DISAGG_HYBRID_FORECAST_
        pd.DataFrame built on previous step
    FORECAST_FLAG
        pd.DataFrame built on previous step
       
    Returns
    -------
    pd.DataFrame
        As a result of this step, a table of the following structure is constructed, T1
    """
    for key in ['product', 'location', 'customer', 'distr_channel']:
        column_name = f"{key}_lvl_id{config_file[f'vf_{key}_lvl']}"
        keys_df = pd.DataFrame(hierarchies[key][column_name]).drop_duplicates()
        FORECAST_FLAG = pd.merge(FORECAST_FLAG, keys_df, 'cross')
    out_time_lvl = 'W'
    if out_time_lvl == 'W':
        if (FORECAST_FLAG['period_start_dt'].apply(lambda x: x - pd.Timedelta(f'{x.dayofweek}D')) != \
            FORECAST_FLAG['period_end_dt'].apply(lambda x: x - pd.Timedelta(f'{x.dayofweek}D'))).any():
                FORECAST_FLAG['out_time_lvl_flg'] = 0
        else:
                FORECAST_FLAG['out_time_lvl_flg'] = 1
#     FORECAST_FLAG['status']=='active'
    FORECAST_FLAG = FORECAST_FLAG.loc[FORECAST_FLAG['status'] == 'active']
    FORECAST_FLAG = pd.merge(FORECAST_FLAG, PRODUCT, on='product_id', how = 'left')
    FORECAST_FLAG = pd.merge(FORECAST_FLAG, LOCATION, on='location_id', how = 'left')
    FORECAST_FLAG = pd.merge(FORECAST_FLAG, CUSTOMER, on='customer_id', how = 'left')
    FORECAST_FLAG = pd.merge(FORECAST_FLAG, DISTR_CHANNEL, on='distr_channel_id', how = 'left')
    
    return FORECAST_FLAG
```

```python
T1 = delete_forecast_inactive_period(DISACC_DISAGG_HYBRID_FORECAST, 
                                     FORECAST_FLAG, 
                                     hierarchies,
                                     PRODUCT, 
                                     LOCATION,
                                     CUSTOMER,
                                     DISTR_CHANNEL)
```
